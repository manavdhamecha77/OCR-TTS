```python
import easyocr

reader = easyocr.Reader(['hi'])  # Hindi
result = reader.readtext('sample-hi.jpeg')

text = " ".join([r[1] for r in result])
print(text)
```

```python
import asyncio
import edge_tts

async def generate_audio(text):
    communicate = edge_tts.Communicate(
        text,
        voice="hi-IN-SwaraNeural"
    )
    await communicate.save("./output-hi.mp3")

await generate_audio(text)
```
